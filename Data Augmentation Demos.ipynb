# Rotatie Aleatorie
      - În conformitate cu documentația Keras la aleatoriu este de 50%

```python
from keras.datasets import mnist
from keras.preprocessing.image import ImageDataGenerator
from matplotlib import pyplot
from keras import backend as k


#incarca baza de date
(x_train,y_train), (x_test,y_test) = mnist.load_data()

#Remodelati datele pentru a fi in format [ samples, width, height, color_depth]
x_train = x_train.reshape(x_train.shape[0], 28,28,1)
x_test = x_test.reshape(x_test.shape[0],28,28,1)

#Schimba tipul de date in float32
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')

#creare un generator de imagine
#Definiti aleatoriu parameti de rotatie la 60 de grade
train_datagen = ImageDataGenerator(rotation_range=60)

#fit parameti din data
train_datagen.fit(x_train)

#Configurati dimensiunea batch si recuperati un batch de imagini
for x_batch, y_batch in train_datagen.flow(x_train, y_train, batch_size=9):
    #creati o grila 3x3 cu imagini
    for i in range(0,9):
        pyplot.subplot(330+1+i)
        pyplot.imshow(x_batch[i].reshape(28,28), cmap=pyplot.get_cmap('gray'))
    #arata plot
    pyplot.show()
    break
```

# Aplica toate in unu

```python
from keras.datasets import mnist
from keras.preprocessing.image import ImageDataGenerator
from keras import backend as k
from matplotlib import pyplot

#Incarcarea baze de date
(x_train, y_train), (x_test, y_test) = mnist.load_data()

#Remodelare a datelor pentru a fi in formatul [ samples, width, height, color_depth]
x_train = x_train.reshape(x_train.shape[0], 28, 28, 1)
x_test = x_test.reshape(x_test.shape[0], 28, 28, 1)

##definim toate in unu
train_datagen = ImageDataGenerator(
    rotation_range=45,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

#fit parametri datelor
train_datagen.fit(x_train)

#configurare dimensiune batch si recuperare un batch de imagini
for x_batch, y_batch in train_datagen.flow(x_train, y_train,9):
    #creati o grila 3x3 cu imagin
    for i in range(0,9):
        pyplot.subplot(330+1+i)
        pyplot.imshow(x_batch[i].reshape(28, 28), cmap=pyplot.get_cmap('gray'))
    #arata plot
    pyplot.show()
    break

```

### Mărirea testului pe o singură imagine
        Ieșiri în directorul ./preview

```python
from keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img

datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

img = load_img('dog.1405.jpg')
x = img_to_array(img) #crearea unui Numpy array cu shape (3, 150, 150)
x = x.reshape((1,) + x.shape) #convetire intr-un numpy array cu shape (1, 3, 150, 150)

i=0
for batch in datagen.flow(x, save_to_dir='output', save_prefix='dog', save_format='jpg'):
    i += 1
    if i > 35:
        break
```
